```python
#load the necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Load the dataset
df = pd.read_csv("dataset.csv")

# Dropping rows with missing height or sex
df_clean = df.dropna(subset=['Height', 'Sex'])

# Group by Year and Sex, and calculate average height
avg_height = df_clean.groupby(['Year', 'Sex'])['Height'].mean().unstack()

# Plotting
plt.figure(figsize=(10, 6), dpi=100)

# Line for Male athletes
plt.plot(avg_height.index, avg_height['M'], 'b^--', label='Male')

# Line for Female athletes
plt.plot(avg_height.index, avg_height['F'], 'ro-', label='Female')

# Title and Labels 
plt.title('Average Athlete Height Over the Years', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 18})
plt.xlabel('Year')
plt.ylabel('Average Height (cm)')

# Add the legend
plt.legend()

# Optional: Save the figure
plt.savefig('athlete_height_trend.png', dpi=300)

# Show plot
plt.show()
```

```python
# Filter only rows with medals and valid regions
medal_df = df.dropna(subset=['Medal', 'region'])

# Group by region and count medals
medal_counts = medal_df['region'].value_counts().head(10)

# Plotting
plt.figure(figsize=(10, 6), dpi=100)
plt.bar(medal_counts.index, medal_counts.values, color='gold', edgecolor='black')

# Titles and labels
plt.title('Top 10 Countries with Most Olympic Medals', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})
plt.xlabel('Country')
plt.ylabel('Number of Medals')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display values on bars
for i, val in enumerate(medal_counts.values):
    plt.text(i, val + 5, str(val), ha='center', fontsize=10)

# Save the chart
plt.savefig('top_countries_medals.png', dpi=300)

# Show plot
plt.show()
```

```python
# Count unique athletes per year
athletes_per_year = df.groupby('Year')['ID'].nunique()

# Plotting
plt.figure(figsize=(10, 6), dpi=100)
plt.plot(athletes_per_year.index, athletes_per_year.values, 'go--', label='Athletes')

# Title and axis labels
plt.title('Total Number of Athletes per Olympic Year', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})
plt.xlabel('Year')
plt.ylabel('Number of Athletes')

# Add grid and legend
plt.grid(True)
plt.legend()

# Show data values on line
for x, y in zip(athletes_per_year.index, athletes_per_year.values):
    plt.text(x, y + 200, str(y), ha='center', fontsize=8)

# Save the plot
plt.savefig('athletes_per_year.png', dpi=300)

# Show the graph
plt.show()
```

```python
# Drop missing age values
age_data = df['Age'].dropna()

# Plotting histogram
plt.figure(figsize=(10, 6), dpi=100)
plt.hist(age_data, bins=30, color='skyblue', edgecolor='black')

# Title and axis labels
plt.title('Distribution of Athlete Ages', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})
plt.xlabel('Age')
plt.ylabel('Number of Athletes')

# Grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Save the figure
plt.savefig('age_distribution_histogram.png', dpi=300)

# Show the plot
plt.show()
```

```python
# Count male and female athletes (drop NaN just in case)
gender_counts = df['Sex'].dropna().value_counts()

# Pie chart
plt.figure(figsize=(8, 8), dpi=100)
plt.pie(gender_counts, 
        labels=gender_counts.index, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['lightblue', 'lightcoral'], 
        wedgeprops={'edgecolor': 'black'})

# Title
plt.title('Gender Distribution of Olympic Athletes', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})

# Save figure
plt.savefig('gender_distribution_pie.png', dpi=300)

# Show chart
plt.show()
```

```python

# Drop rows without a medal
medals = df['Medal'].dropna()

# Count each medal type
medal_counts = medals.value_counts()

# Pie chart
plt.figure(figsize=(8, 8), dpi=100)
plt.pie(medal_counts,
        labels=medal_counts.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=['gold', 'peru', 'silver'],  # Bronze is close to 'peru'
        wedgeprops={'edgecolor': 'black'})

# Title
plt.title('Olympic Medal Distribution by Type', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 16})

# Save the chart
plt.savefig('medal_distribution_pie.png', dpi=300)

# Show plot
plt.show()
```

```python
# Keep only rows with Age and Medal info
df_clean = df.dropna(subset=['Age', 'Medal'])

# Plotting boxplot
plt.figure(figsize=(8, 6), dpi=100)
df_clean.boxplot(column='Age', by='Medal', grid=True, patch_artist=True,
                 boxprops=dict(facecolor='lightgreen', color='green'),
                 medianprops=dict(color='red'),
                 whiskerprops=dict(color='black'),
                 capprops=dict(color='black'),
                 flierprops=dict(markerfacecolor='orange', marker='o', markersize=5, linestyle='none'))

# Titles and labels
plt.title('Athlete Age Distribution by Medal Type')
plt.suptitle('')  # Remove default matplotlib title
plt.xlabel('Medal Type')
plt.ylabel('Age')

# Save the plot
plt.savefig('age_by_medal_boxplot.png', dpi=300)

# Show the plot
plt.show()
```
